```
import osmnx as ox
import matplotlib.pyplot as plt
```

```
place = "Brooklyn, New York"
graph = ox.graph_from_place(place, network_type="drive")
```

```
latlon = (40.693173, -73.986997)
blds = ox.footprints_from_point(latlon, dist=500)
```

```
blds["height"] = blds["height"].astype(float)
blds["height"].fillna(0.0, inplace=True)
```

```
plt.hist(blds.height, 100)
```

```
fig, ax = plt.subplots(figsize=(7, 7))
ax.axis("equal")

blds.plot("height", ax=ax, vmax=100)
```
